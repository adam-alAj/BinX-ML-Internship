## Import necessary libraries

```python
import pandas as pd
import numpy as np

print(f"Pandas Version: {pd.__version__}")
print(f"NumPy Version: {np.__version__}")
```

## 1. Core Pandas Data Structures

```python
# create pandas series
data_series = pd.Series([100,200, 300, 400], index =['Q1', 'Q2', 'Q3', 'Q4'], name="Revenue")
print("--- Pandas Series ---")
print(data_series)
print("\n")

# create a dataFrame from dictionary
raw_data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance', 'IT'],
    'Age': [25, 30, 35, np.nan, 22, 40, 29],
    'Salary': [50000, 75000, 82000, 65000, np.nan, 90000, 72000],
    'Experience_Years': [2, 5, 8, 4, 1, 12, np.nan]
}
df= pd.DataFrame(raw_data)
print("--- DataFrame ---")
df
```

## 2. Data Inspection & Overview

Before analyzing data, we inspect its shape, column types, missing values, and statistical properties using built-in methods like `.head()`, `.info()`, and `.describe()`.

```python
# Display first 5 rows
print("--- First 5 Rows ---")
print(df.head())

# Display DataFrame information (column types, non-null counts)
print("\n--- DataFrame Information ---")
print(df.info())

# Summary statistics for numerical columns
print("\n--- Descriptive Statistics ---")
print(df.describe())
```

## 3. Data Selection & Boolean Filtering

Pandas provides powerful indexing tools:
- **Column Selection**: Accessing single or multiple columns.
- **`.loc`**: Label-based indexing (selects rows and columns by labels or boolean conditions).
- **`.iloc`**: Integer-position based indexing (selects by numeric position).
- **Boolean Masking**: Filtering rows based on logical conditions.
- 

```python
# 1. Selecting specific columns
names_and_salaries = df[['Name', 'Salary']]
print("--- Selected Columns ---")
print(names_and_salaries.head(3))

# 2. Filtering rows where Salary > 70,000 using Boolean Indexing
high_earners = df[df['Salary'] > 70000]
print("\n--- High Earners (Salary > 70,000) ---")
print(high_earners)

# 3. Using .loc with multiple conditions (IT department AND Salary > 70,000)
it_high_earners = df.loc[(df['Department'] == 'IT') & (df['Salary'] > 70000), ['Name', 'Department', 'Salary']]
print("\n--- IT High Earners via .loc ---")
print(it_high_earners)

# 4. Using .iloc to slice the first 3 rows and first 4 columns
sliced_data = df.iloc[0:3, 0:4]
print("\n--- Sliced Data via .iloc ---")
print(sliced_data)
```

## 4. Data Cleaning & Handling Missing Values

Data cleaning involves:
1. Detecting missing values (`NaN`).
2. Imputing missing values using techniques like mean/median filling.
3. Dropping unrecoverable null values.
4. Removing duplicate records.

```python
# Check count of missing values per column
print("--- Missing Values Count Before Cleaning ---")
print(df.isnull().sum())

# Make a copy of the original dataframe for cleaning
df_clean = df.copy()

# 1. Fill missing 'Age' values with the median age (without using inplace=True)
median_age = df_clean['Age'].median()
df_clean['Age'] = df_clean['Age'].fillna(median_age)

# 2. Fill missing 'Salary' values with the mean salary of their respective department
# If a department has all NaNs or single NaN, fill remaining with overall mean
df_clean['Salary'] = df_clean.groupby('Department')['Salary'].transform(lambda x: x.fillna(x.mean()))
df_clean['Salary'] = df_clean['Salary'].fillna(df_clean['Salary'].mean())

# 3. Fill missing 'Experience_Years' with 0
df_clean['Experience_Years'] = df_clean['Experience_Years'].fillna(0)

print("\n--- Cleaned DataFrame (No Missing Values) ---")
print(df_clean)

print("\n--- Missing Values Count After Cleaning ---")
print(df_clean.isnull().sum())
```

## 5. Data Grouping & Aggregation

The `groupby` operation splits the dataset into groups based on specified criteria, applies functions (like sum, mean, count) to each group, and combines the results.


```python
# 1.Average salary per dept
avg_salary_per_dept= df_clean.groupby('Department')['Salary'].mean().reset_index()
print("--- Average salary per dept --- ")
print(avg_salary_per_dept)

# 2.Multiple statistics per group
dept_summary = df_clean.groupby('Department').agg(
    Employee_count=('Employee_ID', 'count'),
    Avg_salary= ('Salary', 'mean'),
    Max_Salary= ('Salary','max'),
    Avg_Experience= ('Experience_Years','mean')
).reset_index()

print("\n--- Detailed Department Summary ---")
print(dept_summary)
```

## 6. Saving Cleaned Data & Summary

```python
# Save clean dataset to CSV inside Day_4 directory
df_clean.to_csv('cleaned_employee_data.csv', index=False)
print("Cleaned data successfully saved to 'cleaned_employee_data.csv'.")
```

# Hands-On Lab: Exploring a Real Dataset with Pandas

## Overview
In this practical lab, we perform an end-to-end exploratory data analysis (EDA) on a real-world supermarket sales dataset directly loaded via URL. The workflow covers:
1. **Dataset Loading & Inspection:** Loading CSV data and reporting shape, columns, and data types.
2. **Data Cleaning & Missing Value Handling:** Identifying missing values and applying justified imputation or dropping strategies.
3. **Subsetting & Filtering:** Extracting meaningful slices of data based on specific conditions.
4. **Group Aggregation:** Computing descriptive metrics grouped by categorical features using `groupby`.

```python
#Official raw CSV dataset from GitHub (Seaborn Official Repository)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"

# Load dataset into Pandas DataFrame
df_lab = pd.read_csv(url)

print("✅ Dataset successfully loaded into DataFrame!\n")

# Report dataset shape
print(f"1. Dataset Shape (Rows, Columns): {df_lab.shape}\n")

# Report list of columns and data types
print("2. Column Names and Data Types:")
print(df_lab.dtypes)

# Preview first 5 rows
print("\n3. First 5 Rows Preview:")
df_lab.head()
```

## Step 2: Handling Missing Values & Data Cleaning

### Documented Justification:
1. **Detection:** We count null values across all columns to locate missing data points.
2. **Imputation Strategy:**
   - Numerical columns with missing values (e.g., `horsepower`) are filled using the **median** value of that column.
   - *Why Median?* The median is robust against extreme outliers, ensuring we do not distort the central tendency of engine power metrics.
3. **Deduplication:** Duplicate rows are checked and dropped to preserve dataset integrity.

```python
# 1. Count missing values per column
print("--- Missing Values per Column before Cleaning---")
print(df_lab.isnull().sum())

# 2. Impute missing values in numeric columns using median
numeric_cols = df_lab.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    if df_lab[col].isnull().sum() > 0:
        median_val = df_lab[col].median()
        df_lab[col] = df_lab[col].fillna(median_val)
        print(f"\n-> Imputed missing values in '{col}' with median: {median_val}")

# 3. check and drop duplicate rows
initial_rows = len(df_lab)
df_lab.drop_duplicates(inplace=True)

print(f"\nDuplicates removed: {initial_rows - len(df_lab)}")
print(f"Total missing values remaining: {df_lab.isnull().sum().sum()}")
```

## Step 3: Data Filtering & Subset Exploration

### Subset Description & Findings:
- **Filtering Threshold:** We filter the dataset to isolate high fuel-efficiency cars where Miles Per Gallon (`mpg`) exceeds **30.0**.
- **Analytical Goal:** Explore the physical properties (cylinders, weight, horsepower) of eco-friendly vehicles compared to the general population.

```python
# filter rows where mpg > 30.0
high_efficiency_cars = df_lab[df_lab['mpg'] > 30.0]

total_count = len(df_lab)
subset_count = len(high_efficiency_cars)
percentage = (subset_count / total_count) * 100

print(f"Total Car Models: {total_count}")
print(f"High Fuel-Efficiency Models (mpg > 30): {subset_count}")
print(f"Percentage of High Efficiency Models: {percentage:.2f}%\n")

# Preview specific attributes of the subset
display_cols = ['name', 'mpg', 'cylinders', 'horsepower', 'weight', 'origin']
print("--- High Fuel-Efficiency Sample Preview ---")
high_efficiency_cars[display_cols].head()
```

## Step 4: Groupby Categorical Aggregation

### Aggregation Metrics:
We group the dataset by manufacturing region (`origin`) to calculate:
1. **`Average_MPG`**: Mean fuel economy per origin.
2. **`Average_Horsepower`**: Average engine output.
3. **`Average_Weight`**: Average vehicle mass.
4. **`Model_Count`**: Number of car models tested from each region.

```python
# Aggregate stats by manufacturing origin
origin_summary = df_lab.groupby('origin').agg(
    Average_MPG=('mpg', 'mean'),
    Average_Horsepower=('horsepower', 'mean'),
    Average_Weight=('weight', 'mean'),
    Model_Count=('name', 'count')
).reset_index()

# Sort by Average MPG descending
origin_summary = origin_summary.sort_values(by='Average_MPG', ascending=False)

print("--- Regional Vehicle Performance Summary ---")
origin_summary
```

##  Final Interpretation of Results

Based on the aggregated metrics grouped by manufacturing region (`origin`), we draw the following conclusions:

1. **Fuel Efficiency Leadership:**
   - **Japan** leads in fuel economy with an average of **30.45 MPG**, closely followed by **Europe** at **27.89 MPG**.
   - **USA** lags behind with an average of **20.08 MPG**.

2. **Power & Weight Trade-off:**
   - **American vehicles** prioritize high performance and size, having the highest average horsepower (**118.64 HP**) and average weight (**3361.93 lbs**).
   - **Japanese and European vehicles** feature lighter builds (~2221 lbs and ~2423 lbs) and moderate power output (~80 HP), which directly contributes to their superior fuel efficiency.

3. **Sample Distribution:**
   - The dataset is predominantly composed of American models (**249 models**), compared to Japanese (**79 models**) and European (**70 models**).

---
###  Summary Conclusion
Lighter vehicle weight and lower engine power output are strong indicators of high fuel efficiency (`MPG`). Japanese manufacturers achieved the best balance between weight reduction and fuel performance in this dataset.
